```python
import pandas as pd

data = pd.read_excel("expense1.xlsx")
df = pd.DataFrame(data)
#print(df)

group = df.groupby("Category").agg({"Amount":"sum"})
print(group)

plt.step(group.index, group["Amount"],where = "mid", marker = "o")
plt.show()

```
